```python
import numpy as np
import matplotlib.pyplot as plt


# Intervalo de x e y no qual se deseja analisar o grfico

xp = np.linspace(-5,5,100)
yp = np.linspace(-5,5,100)
X,Y = np.meshgrid(xp,yp)

# Z = f(X,Y) base

#Z = 10 - 4*X - 5*Y

Z = Y*np.exp(-X)

# Plot das curvas de nvel

plt.contour(X,Y,Z,10,cmap='jet')
plt.colorbar()
plt.show()

#Plot da superfcie 3D

plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
mysurf = ax.plot_surface(X,Y,Z,cmap='jet',alpha=1)
#plt.colorbar(mysurf)
#cset = plt.contourf(X,Y,Z,zdir='z',offset=0,cmap='jet')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
